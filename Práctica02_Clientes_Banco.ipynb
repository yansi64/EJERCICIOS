```python
import simpy
import random

# Proceso de atención de un cliente
def cliente(env, nombre, cajero):
    print(f"{nombre} llega al banco en el minuto {env.now}")
    with cajero.request() as req:
        yield req
        print(f"{nombre} comienza a ser atendido en el minuto {env.now}")
        tiempo_atencion = random.randint(3, 7)  # tiempo de servicio
        yield env.timeout(tiempo_atencion)
        print(f"{nombre} termina en el minuto {env.now}")

# Generador de clientes
def generador_clientes(env, cajero):
    i = 0
    while True:
        yield env.timeout(random.randint(2, 5))  # tiempo entre llegadas
        i += 1
        env.process(cliente(env, f"Cliente {i}", cajero))

# Configuración de la simulación
env = simpy.Environment()
cajero = simpy.Resource(env, capacity=1)  # solo un cajero
env.process(generador_clientes(env, cajero))

# Ejecutar simulación por 30 minutos
env.run(until=30)

#Explicación del Código
#simpy.Environment() crea el entorno de simulación.
#Resource representa el recurso limitado (el cajero).
#Cada cliente espera si el cajero está ocupado.
#La simulación corre durante 30 minutos simulados.
```
